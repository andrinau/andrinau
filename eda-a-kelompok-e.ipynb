```python
# Importing Some Python Modules
import warnings; warnings.simplefilter('ignore')
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt, seaborn as sns
from collections import Counter

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
bm = pd.read_csv("../input/ashrae-energy-prediction/building_metadata.csv")
bm
```

```python
train = pd.read_csv("../input/ashrae-energy-prediction/train.csv")
train
```

```python
w_train = pd.read_csv("../input/ashrae-energy-prediction/weather_train.csv")
w_train
```

```python
gb_train = bm.merge(train, how = "left", on = ["building_id"]).merge(w_train, how = "left", on = ["site_id","timestamp"]) #melakukan penggabungan
gb_train
```

Penggunaan merge untuk menggabungkan kolom, lalu penggunaan how = left untuk memasukan baris pada data kanan yang sama dengan data kiri ke data kiri, sehingga baris pada data kiri bertambah

```python
gb_train.describe() # Statistika deskriptif dari data "Numerik"
```

penggunaan describe untuk melihat hasil statistika deskriptif pada gb_train dimana terdapat mean,median, q1,q2,q3 dll

```python
gb_train.info() # tipe data di setiap kolom
```

penggunaan info digunakan untuk mengetahui tipe apa saja yang terdapat dalam data gb_train

```python
gb_train["site_id"] = gb_train["site_id"].astype("category")
gb_train["building_id"] = gb_train["building_id"].astype("category")
gb_train["primary_use"] = gb_train["primary_use"].astype("category")
gb_train["meter"] = gb_train["meter"].astype("category")
gb_train["timestamp"] = gb_train["timestamp"].astype("category") 
gb_train["year_built"] = gb_train["year_built"].astype("category")
gb_train.dtypes # tipe data di setiap kolom
```

penggunaan .astype() digunakan untuk data kategorik yang masih bertipe float, int dll ke tipe category untuk mempermudah dalam preprosesing

```python
gb_train2 = gb_train[["meter","primary_use","meter_reading","square_feet","air_temperature"]] #mengambil kolom secara manual
gb_train2
```

penggunaan kode ini digunakan untuk mengambil variabel meter, primary use, meter reading, square feet dan air temperature secara manual agar preprosesing lebih mudah

```python
gb_train2.isnull().sum() # General Look at the Missing Values
```

penggunaan isnull.sum untuk melihat dan menghitung jumlah dari variabel mana saja yang memiliki missing value

```python
(gb_train2.isnull().sum()/len(gb_train2)*100).to_frame('persentase missing') # General Look percentase at the Missing Values
```

penggunaan kode ini di gunakan untuk menghitung persentase dari jumlah missing value yang telah kita lihat
